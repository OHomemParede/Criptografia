### Entrando com a imagem
#### uma senha é gerada com o mesmo tamanho da stream de bits da imagem 

```python
image_name = 'imagens/ada.jpg'#input('Nome da imagem com extensão: ') # entrando com a imagem
image, altura, largura, canais = input_img(image_name) # carregando a imagem e pegando suas dimensões
show_image(image) # mostrando a imagem
```

---------------------

### Operação AND

```python
ENC = criptografar(image, altura, largura, canais,'AND') # ENC é uma stream de bits

AND_IMAGE = image_builder(altura, largura, canais, ENC)
show_image(AND_IMAGE)
```

-------------------------------

### Operação OR    

```python
ENC = criptografar(image, altura, largura, canais,'OR') # ENC é uma stream de bits

OR_IMAGE = image_builder(altura, largura, canais, ENC)
show_image(OR_IMAGE)
```

------------------

### Operação XOR 

```python
ENC = criptografar(image, altura, largura, canais,'XOR') # ENC é uma stream de bits

XOR_IMAGE = image_builder(altura, largura, canais, ENC)
show_image(XOR_IMAGE)
```
